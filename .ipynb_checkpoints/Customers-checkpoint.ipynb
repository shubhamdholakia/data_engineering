```python
import pandas as pd
cust_pflo = pd.read_csv("Customer Portfolio Snapshot.csv")
```

```python
cust_pflo.head()
```

```python
cust_pflo.tail()
```

```python
cust_demo = pd.read_csv("Customer Demographics.csv")
cust_demo.head(10)
```

```python
cust_investment = pd.read_csv("Customer Investment Snapshot.csv")
cust_investment.head()
cust_investment.duplicated()
```

```python
cust_attrn = pd.read_csv("Customer Attrition Status.csv")
cust_attrn.head()
```

```python
cust_pflo.duplicated()
```

```python
pflo_demo_join = pd.merge(cust_pflo,cust_demo,on = ['RowNumber','CustomerId'],how = 'left')
pflo_demo_join.head()
```

```python
pflo_demo_join.duplicated()
```

```python
pflo_demo_join.isnull().sum()
```

```python
pflo_demo_join.isnull().tail()
```

```python
# we have joined customer portfolio and respective demographics . We have found that some customers in our portfolia dont have a
# checking account , have no products , and no checking account and no details in customer demographics too .... 
# we wont take these customers in consideration for our analysis ... dropping them in our joined table .. 
```

```python
pflo_demo_join.dropna(subset = ['IsActiveMember', 'HasChckng'],inplace = True)
# to check table stats used describe()
pflo_demo_join.describe()
pflo_demo_join.count()
pflo_demo_join.isnull().sum()
pflo_demo_join.isnull().tail()
pflo_demo_join.tail()
pflo_demo_join.loc[pflo_demo_join.duplicated(subset = ['CustomerId']),:]
pflo_demo_join.loc[pflo_demo_join.duplicated(keep = 'first'),:]
pflo_demo_join.drop_duplicates(keep = 'first',inplace = True)
pflo_demo_join.loc[pflo_demo_join.duplicated()]
pflo_demo_join.describe()
```

```python
pflo_demo_investment_join = pd.merge(pflo_demo_join,cust_investment,on = ['RowNumber','CustomerId'],how = 'left')
pflo_demo_investment_join.head()
```

```python
pflo_demo_investment_join.duplicated().sum()
```

```python
# Here there are a lot of balance columns that are null .... There are not any 0 balance accounts so for guestimate we are assuming
# that for 0 balance accounts are o/p as null in the file .. However since this is a guestimate ... we will keep this value as 10
pflo_demo_investment_join.isnull().sum()
pflo_demo_investment_join['Balance'].fillna(10,inplace = True)
pflo_demo_investment_join.head()
```

```python
pflo_demo_investment_join.isnull().sum()
```

```python
pflo_demo_investment_join.describe()
```

```python
# CreditScore is very important in banking ... There must be a formula in place however here simply by applying mean at places 
# where it is null
pflo_demo_investment_join['CreditScore'].fillna(650,inplace = True)
```

```python
pflo_demo_investment_attr_join = pd.merge(pflo_demo_investment_join,cust_attrn,on = ['RowNumber','CustomerId'])
```

```python
result = pflo_demo_investment_attr_join.sort_values("RowNumber").copy()
```

```python
result.reset_index( inplace = True,drop = True)
```

```python
result
```

```python
result = result[['RowNumber','CustomerId','Surname','CreditScore','Geography','Gender','Age','Tenure','Balance','NumOfProducts'
                ,'HasChckng','IsActiveMember','EstimatedSalary','Exited']]
```

```python
result.duplicated().sum()
```

```python
geography = result.groupby(['Geography','Exited'])
```

```python
df = pd.DataFrame(geography['Geography','Exited'].count(),columns = ['Exited'])
df.head(6)
```
